6. การป้องกัน Data Leakage ในกรณีนี้
- ทำ Scaling และ Encoding(one-hot-encoding) โดยใช้ข้อมูลจากชุดฝึก (Training set)เท่านั้น 
แล้วนำค่าพารามิเตอร์ที่ได้ไปใช้กับชุดทดสอบ
    - Numerical Features 
    จำเป็นต้องผ่านกระบวนการ Scaling เพื่อปรับขนาดข้อมูลให้อยู่ในช่วงเดียวกัน โดยใช้ StandardScaler จากsklearn 
    เพื่อให้ค่ามีค่าเฉลี่ยเป็น 0 และส่วนเบี่ยงเบนมาตรฐานเป็น 1 ซึ่งช่วยให้โมเดลไม่ถูก bias จากฟีเจอร์ที่มีค่ามากเกินไป
    - Categorical Features
    ฟีเจอร์ที่เป็นข้อความ ไม่สามารถนำไปใช้กับโมเดลได้โดยตรง จึงต้องแปลงเป็นตัวเลขผ่านกระบวนการ One-Hot Encoding
    โดยใช้ OneHotEncoder จาก sklearn เพื่อแปลงแต่ละค่าหมวดหมู่เป็นคอลัมน์ใหม่ที่มีค่า 0 หรือ 1 ซึ่งช่วยให้โมเดลเข้าใจข้อมูลโดยไม่สับสนเรื่องลำดับ
- ตรวจสอบให้แน่ใจว่าไม่มีการใช้ฟีเจอร์ที่สัมพันธ์กับผลลัพธ์ หรือกับข้อมูลในอนาคตที่ไม่ควรปรากฏในการฝึกโมเดล 

7. วิเคราะห์ความสำคัญของคุณลักษณะ ตัวไหนควรและไม่ควรใช้

```python
from sklearn.linear_model import LinearRegression
from sklearn.inspection import permutation_importance

df_train_numeric = df_employee_train.select_dtypes(include=["int64", "float64"]).dropna()
df_test_numeric = df_employee_test.select_dtypes(include=["int64", "float64"]).dropna()

target_column = "MonthlyIncome"
X_train = df_train_numeric.drop(target_column, axis=1)
y_train = df_train_numeric[target_column]

X_test = df_test_numeric.drop(target_column, axis=1)
y_test = df_test_numeric[target_column]

model = LinearRegression()
model.fit(X_train, y_train)

result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
```

```python
result_df = pd.DataFrame({
    "feature": X_test.columns,
    "score": result.importances_mean
})
result_df
```
