```python
import pandas as pd
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
selected_columns = [
    "Education",
    "Gender",
    "MaritalStatus",
    "NumCompaniesWorked",
    "TrainingTimesLastYear",
    "YearsWithCurrManager",
    "TotalWorkingYears",
    "EmployeeCount", 
    "Over18", 
    "StandardHours", 
    "EmployeeNumber"
]

data = df.drop(columns=selected_columns)

data.head()
```

```python
data.rename(columns={
    "BusinessTravel": "TravelFrequency",
    "EducationField": "EducationMajor",
    "YearsAtCompany": "YearsInCompany",
    "NumCompaniesWorked": "PreviousCompaniesWorked"
}, inplace=True)
```

```python
data.head()
```

```python
data['Attrition'] = data['Attrition'].map({'Yes': True, 'No': False})
data['OverTime'] = data['OverTime'].map({'Yes': True, 'No': False})
```

```python
data['TravelFrequency'] = data['TravelFrequency'].astype('category')
data['Department'] = data['Department'].astype('category')
```
