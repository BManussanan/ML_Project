4. ข้อมูลมีการ/น่าจะมีการกำหนดเฉลย (labeling) ด้วยวิธีใด เพราะเหตุใด มีปัญหาหรือไม่ อย่างไร แก้ไขได้อย่างไร 

ข้อมูลชุดนี้มีการกำหนดเฉลย (labeling) ด้วยวิธี explicit labeling โดยใช้คอลัมน์ Attrition ซึ่งแสดงสถานะว่าพนักงานลาออกหรือไม่ แต่ตามที่เว็บไซต์ต้นทางระบุ ข้อมูลชุดนี้ถูกสร้างขึ้นโดยนักวิจัยเพื่อใช้สำหรับการทดลองโมเดลในงานด้านทรัพยากรบุคคล (HR Analytics) ดังนั้นการกำหนดเฉลยในลักษณะนี้ถือเป็น explicit labeling + hand labeling เพราะ นักวิจัยเป็นผู้ออกแบบกฎหรือความสัมพันธ์ของข้อมูล แล้วกำหนดว่าในกรณีใดควรให้ค่า Attrition เป็น “Yes” หรือ “No” เช่น หากพนักงานมีรายได้ต่ำ ทำงานล่วงเวลาบ่อย และมีระดับความพึงพอใจต่ำ อาจถูกกำหนดให้ลาออก

ปัญหา:

- ข้อมูลอาจไม่สะท้อนพฤติกรรมจริงของพนักงานในโลกความเป็นจริง

- ความสัมพันธ์ระหว่างฟีเจอร์กับ label อาจถูกกำหนดไว้ชัดเจนเกินไป ทำให้โมเดลเรียนรู้ได้ง่ายกว่าความเป็นจริง

การแก้ไข:

- ใช้ข้อมูลจริงควบคู่กับข้อมูลจำลองเพื่อฝึกและทดสอบโมเดล

- หากใช้ข้อมูลจำลอง ควรทดสอบโมเดลกับชุดข้อมูลจริงในภายหลัง

- พิจารณาปรับหรือสุ่มเงื่อนไขการลาออกให้หลากหลายขึ้น เพื่อเพิ่มความสมจริง 


5. ข้อมูลขาดความสมดุล (imbalance) หรือไม่ อย่างไร มากน้อยแค่ไหน มีปัญหาหรือไม่ อย่างไร แก้ไขได้อย่างไร

มีความไม่สมดุล(Imbalanced Classes) โดยเฉพดาะในตัวแปร Attriton ที่แสดงว่าพนักงงานลาออก หรือไม่ลาออก 
- พนักงานกลุ่มลาออก มีจำนวนน้อยกว่ามาก
- Model จะเรียนรู้กลุ่มลาออกได้ยาก เพราะ ข้อมูลน้อย ทำให้พลาดเป้ากลุ่มเป้าหมายจริง(ลาออก)
- นำไปสู่ Asymmetric Costs of Error 

วิธีแก้ไขปัญหา
Data-level Methods:
- Oversampling: เพิ่มข้อมูลกลุ่มลาออก ด้วยเทคนิค SMOTE(เทคนิคเพิ่มข้อมูลในกลุ่มที่มีน้อย (minority class) โดยสร้างตัวอย่างใหม่จากข้อมูลจริง แทนที่จะก็อปปี้แบบซ้ำๆ)
- Undersampling: ลดจำนวนข้อมูลกลุ่มไม่ลาออก ช่วยให้ balance ดีขึ้นก่อนฝึกโมเดล

Algorithm-level Methods:
- Class-balanced loss: คำนวณน้ำหนักจากจำนวน class อย่างชาญฉลาด
- Focal Loss: ปรับ loss ให้เน้น sample ที่ยากขึ้น (เช่น การลาออก)

6. จำเป็นหรือสามารถสร้างข้อมูลเพิ่ม (augmentation) ได้หรือไม่ อย่างไร

- ตัวอย่างกลุ่มที่สนใจมีน้อยมาก
- การเพิ่มข้อมูลช่วยให้โมเดลไม่ Overfit กลุ่มใหญ่ และเรียนรู้ลักษณะของกลุ่มน้อยได้ดีขึ้น
